```python
# We will use pearson corelation
plt.figure(figsize=(18,18))
sns.heatmap(df.corr(),annot=True,cmap=plt.cm.CMRmap_r)
```

here we can see RAD and TAX are highly correlated, ao we will take only one feature

below we will create a function to find the columns that are corelated

```python
def correlate(dataset,threshold):
    col_cor= set()
    cor_mat = dataset.corr()
    for i in range(len(dataset.columns)):
        for j in range (i):
            if abs(cor_mat.iloc[i,j])> threshold:
                col_cor.add(cor_mat.columns[i])
    return col_cor
            
```

```python
print(correlate(df,0.7))
```

from the above we can see that having threshold value 0.7, 5 columns are corelated. so we can remove those columns
