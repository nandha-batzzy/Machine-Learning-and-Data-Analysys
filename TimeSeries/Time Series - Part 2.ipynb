```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('AAPL.csv',parse_dates=['Date'])
```

```python
data
```

```python
data.drop('Date',axis=1,inplace=True)
```

```python
data
```

```python
## if we need to get the dates for the above dataframe we use daterange

rang = pd.date_range(start="10/02/2017",end="08/10/2021",freq='B')
```

```python
rang
```

from the above we can see it gives the range based on the freaquency type we gave which is 'B'

```python
### now we can set index as rang
data.set_index(rang,inplace=True)
```

```python
data
```

now we can see the date index have been replaced

```python
data.Close.plot()
```

```python
## asfreq will change the freaquency of the data frame index 
# if you use pad,the week ends gap will be filled with the before date records
data.asfreq('D',method='pad').head(10)
```
